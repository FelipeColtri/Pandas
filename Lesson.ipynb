```python
chicago_crimes[chicago_crimes['Location Description'] == 'APARTMENT'] # all rows where "Location Description" is "APARTMENT"
```

```python
chicago_crimes['Location Description'] == 'APARTMENT' # inner command above (boolean)
```

```python
# condition AND => '&', OR => '|' (parentheses are required)
chicago_crimes[(chicago_crimes['Location Description'] == 'APARTMENT') & (chicago_crimes['Ward'] > 15.0)]
```

```python
# not equal '!=' operator and OR condition
chicago_crimes[(chicago_crimes['Location Description'] != 'APARTMENT') | (chicago_crimes['Ward'] > 15.0)]
```

```python
# show "Location Description" is 'RESIDENCE' OR 'APARTMENT' OR 'STREET'
chicago_crimes[(chicago_crimes['Location Description'] == 'RESIDENCE') | (chicago_crimes['Location Description'] == 'APARTMENT') | (chicago_crimes['Location Description'] == 'STREET')]
```

```python
# same result as above using function 'isin()' (don't forget the square brackets)
chicago_crimes[chicago_crimes['Location Description'].isin(['RESIDENCE', 'APARTMENT', 'STREET'])]
```

```python
# combination of 'isin()' with AND condition and '<=' operator
chicago_crimes[chicago_crimes['Location Description'].isin(['RESIDENCE', 'APARTMENT', 'STREET']) & (chicago_crimes['Ward'] <= 15.0)]
```

```python
chicago_crimes[chicago_crimes['Location'].isna()] # select rows where 'Location' IS empty
```

```python
chicago_crimes[chicago_crimes['Location'].notna()] # select rows where 'Location' IS NOT empty
```
